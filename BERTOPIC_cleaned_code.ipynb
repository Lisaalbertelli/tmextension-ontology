Dataset Splitting (ItaParlCorpus):

Splitting of the examinated ItaParlCorpus datasets into smaller subfiles for improved manageability, covering period: 1948 - 2006

To improve processing efficiency, the corpus data from 1948 through 2005 is being split into smaller files.

The core data used for this thesis originates from the ItaParlCorpus, specifically utilizing the following three files retrieved from the Harvard Dataverse 
(DOI: 10.7910/DVN/KUARWD): 
camera\_1948-1972.csv, camera\_1972\_1992.csv, and camera\_1992-2006.csv. 

To improve manageability and processing efficiency, these datasets, collectively spanning the period from 1948 to 2006, were initially partitioned into smaller subfiles. Subfiles created from the first dataset (camera\_1948-1972.csv) are sequentially named as part1, part2, part3, and so forth. 

Subfiles derived from the second dataset (camera\_1972\_1992.csv) follow the same numerical naming pattern but include a trailing 'b' (e.g., part1b) to signal their origin from the second considered dataset. 

Similarly, subfiles from the final dataset (camera\_1992-2006.csv) use the same numerical pattern with a trailing 'c' (e.g., part1c) to identify them as belonging to the last dataset.

Preprocessing

This script consolidates all split subfiles into a single DataFrame, while selectively loading essential columns and performing immediate data cleaning by filtering out procedural speech (Chair = True) 

```python
#unisco i sottofile in un unico dataframe - per limite di ram il mio df unisce tutti i sottofiles considerando solo le colonne doc_id, speaker e text (+chair)
import pandas as pd

# Colonne da caricare. 'chair' è necessaria per il filtro, ma la rimuoveremo dopo.
cols_to_use = ['doc_id', 'speaker', 'text', 'chair', 'pageid_wiki', 'row_id' ]

# Lista di tutti i percorsi dei file
file_paths = [
    r"C:\Users\lisas\Desktop\ItaParl 1948-1972\parte_1.csv",
    r"C:\Users\lisas\Desktop\ItaParl 1948-1972\parte_2.csv",
    r"C:\Users\lisas\Desktop\ItaParl 1948-1972\parte_3.csv",
    r"C:\Users\lisas\Desktop\ItaParl 1948-1972\parte_4.csv",
    r"C:\Users\lisas\Desktop\ItaParl 1948-1972\parte_5.csv",
    r"C:\Users\lisas\Desktop\ItaParl 1948-1972\parte_6.csv",
    r"C:\Users\lisas\Desktop\ItaParl 1948-1972\parte_7.csv",
    r"C:\Users\lisas\Desktop\ItaParl 1972-1992\parte_1b.csv",
    r"C:\Users\lisas\Desktop\ItaParl 1972-1992\parte_2b.csv",
    r"C:\Users\lisas\Desktop\ItaParl 1972-1992\parte_3b.csv",
    r"C:\Users\lisas\Desktop\ItaParl 1972-1992\parte_4b.csv",
    r"C:\Users\lisas\Desktop\ItaParl 1972-1992\parte_5b.csv",
    r"C:\Users\lisas\Desktop\ItaParl 1972-1992\parte_6b.csv",
    r"C:\Users\lisas\Desktop\ItaParl 1972-1992\parte_7b.csv",
    r"C:\Users\lisas\Desktop\ItaParl 1992-2006\parte_1c.csv",
    r"C:\Users\lisas\Desktop\ItaParl 1992-2006\parte_2c.csv",
    r"C:\Users\lisas\Desktop\ItaParl 1992-2006\parte_3c.csv",
    r"C:\Users\lisas\Desktop\ItaParl 1992-2006\parte_4c.csv",
    r"C:\Users\lisas\Desktop\ItaParl 1992-2006\parte_5c.csv",
    r"C:\Users\lisas\Desktop\ItaParl 1992-2006\parte_6c.csv",
    r"C:\Users\lisas\Desktop\ItaParl 1992-2006\parte_7c.csv",

]

clean_dfs = []

# Ciclo su ogni file
for path in file_paths:
    # La logica di codifica ora è gestita da try-except per essere più robusta
    try:
        # 1. Prova a leggere il file con la codifica standard 'utf-8'
        print(f"Tentativo di lettura di {path} con codifica 'utf-8'...")
        df_temp = pd.read_csv(path, sep=';', usecols=cols_to_use, encoding='utf-8')
    except UnicodeDecodeError:
        # 2. Se fallisce con un errore di decodifica, prova con 'latin1'     #questo è necessario perchè a partire dal file 7b, la codifica non è più utf-8 ma è latin1
        print(f"Fallito. Tentativo con codifica 'latin1'...")
        df_temp = pd.read_csv(path, sep=';', usecols=cols_to_use, encoding='latin1')


    
    # --- PASSO 1: Rimuovi le righe procedurali: righe in cui chair = true e speaker = president
    # Usa ~ per invertire la condizione: tieni tutto tranne le righe procedurali
    df_step1 = df_temp[~((df_temp['chair'] == "TRUE") | (df_temp['speaker'] == "PRESIDENTE")) & df_temp['text'].notna()]
    
    # --- PASSO 2: Dopo aver rimosso gli interventi procedurali, ora elimino tutte quelle righe in cui text = < 5 tokens tutti UPPERCASE. Questo perchè molte righe avevano solo un nome proprio, uppercase e sporcavano i topics
    mask_short_uppercase = (
        (df_step1['text'].str.split().str.len() < 5) &
        (df_step1['text'].apply(lambda x: all(t.isupper() for t in x.split()) if pd.notna(x) else False))
    )
    df_clean = df_step1[~mask_short_uppercase]
    
    clean_dfs.append(df_clean)

# Concatena tutto e rimuovi la colonna 'chair'
df = pd.concat(clean_dfs, ignore_index=True)
df = df.drop(columns=['chair'])

print("Operazione completata. Il DataFrame 'df' è pronto.")
print(f"Dimensione finale: {df.shape}")
```

```python
#This step generates a unique identifier (id_custom) for each speaking turn. 
# It  combines the existing doc_id with the pageid_wiki value if the latter is available; 
# otherwise, it uses the doc_id paired with the speaker name as a fallback identifier.

df['id_custom'] = df.apply(
    lambda row: f"{row['doc_id']}||{row['pageid_wiki']}" if pd.notna(row['pageid_wiki']) and str(row['pageid_wiki']).strip() != "" 
    else f"{row['doc_id']}||{row['speaker']}", 
    axis=1
)
```

```python
# the data is aggregated by this custom key (doc_key). 
# This creates a final DataFrame (doc_df) where each row represents a distinct speech event, with all associated cleaned text segments concatenated.

#  Aggrega per id_custom (doc_id + pageid_wiki/speaker)
df['doc_key'] = df['id_custom']

doc_df = df.groupby('doc_key').agg({
    'doc_id': 'first',
    'pageid_wiki': 'first',
    'speaker': 'first',
    'text': lambda texts: " ".join(texts)
}).reset_index(drop=False)
```

```python
import pandas as pd
import re

# Assicurati che 'doc_df' sia caricato in memoria.
# Se hai appena eseguito lo script precedente, dovrebbe esserlo.
# Se no, ricaricalo da:
# doc_df = pd.read_csv(r"C:\Users\lisas\Desktop\ItaParl 1948-1972\new_id_texts.csv")

try:
    # Definizione della funzione per rimuovere il nome del parlante
    def remove_speaker_name_completely(row):
        """
        Rimuove ogni occorrenza del nome del parlante dal testo in modo robusto.
        Gestisce diversi formati del nome (intero, singoli token) ed è case-insensitive.
        """
        # Usa la colonna 'text' che abbiamo creato e pulito in precedenza
        text = row['text']
        speaker = row['speaker']

        # Controlli di sicurezza per evitare errori se i dati non sono validi
        if pd.isna(text) or pd.isna(speaker) or not isinstance(speaker, str) or not speaker.strip():
            return text

        # 1. Rimuovi ogni singolo token del nome (case-insensitive e solo parole intere)
        # Questo è il passo più importante. Gestisce casi come "Giovanni ha detto... e poi Spadolini..."
        speaker_tokens = speaker.lower().split()
        for token in speaker_tokens:
            # \b (word boundary) è cruciale: rimuove "Marco" ma non "marcatura"
            # re.escape() gestisce nomi con caratteri speciali
            text = re.sub(r'\b' + re.escape(token) + r'\b', '', text, flags=re.IGNORECASE)

        # 2. Rimuovi il nome completo come stringa (case-insensitive) per catturare occorrenze concatenate
        # Questo è un passo aggiuntivo per maggiore pulizia
        speaker_lower = speaker.lower()
        text = text.replace(speaker_lower, '')

        # 3. Pulisci gli spazi multipli che si sono creati dopo le rimozioni
        text = re.sub(r'\s+', ' ', text).strip()
            
        return text

    # Applica la funzione a ogni riga del DataFrame per creare la nuova colonna
    # axis=1 indica che la funzione deve essere applicata alle righe
    print("Applicazione della funzione di rimozione del nome in corso...")
    doc_df['embedding_text'] = doc_df.apply(remove_speaker_name_completely, axis=1)
    print("Funzione applicata con successo.")

    # Visualizziamo un esempio per verificare il risultato
    print("\n--- Verifica della Rimozione del Nome ---")
    # Selezioniamo lo stesso discorso di prima per confronto
    # (Assicurati che la variabile 'doc_id' dall'esempio precedente sia ancora disponibile o selezionane uno nuovo)
    try:
        # Riusiamo lo stesso ID di prima se disponibile, altrimenti ne prendiamo uno nuovo
        example_doc_id_to_check = doc_id 
    except NameError:
        random_speech_for_check = doc_df.sample(n=1, random_state=99)
        example_doc_id_to_check = random_speech_for_check['doc_id'].iloc[0]

    example_row = doc_df[doc_df['doc_id'] == example_doc_id_to_check].iloc[0]

    print(f"Controllo per il Doc ID: {example_doc_id_to_check}")
    print(f"Parlante: {example_row['speaker']}")
    print("-" * 20)
    print("TESTO ORIGINALE (dalla colonna 'text'):")
    print(example_row['text'][:300] + "...") # Mostriamo i primi 300 caratteri
    print("-" * 20)
    print("TESTO PULITO (dalla colonna 'embedding_text'):")
    print(example_row['embedding_text'][:300] + "...")
    print("-" * 20)


except NameError:
    print("Errore: Il DataFrame 'doc_df' non è definito.")
    print("Per favore, esegui prima gli script di consolidamento e pulizia dei dati.")
except Exception as e:
    print(f"Si è verificato un errore: {e}")
```

Obtaining cleaned text that will be used after calculating embeddings to define the vectorizer (CountVectorizer)

```python
# ===================================================================
# CELLA 1: IMPOSTAZIONE DI SPACY E DEFINIZIONE DELLE FUNZIONI
# ===================================================================

# Installa il modello linguistico italiano per spaCy se non è già presente
!pip install -q it_core_news_sm

import spacy
from spacy.lang.it.stop_words import STOP_WORDS
import pandas as pd
import re

# Carica il modello linguistico italiano "small" di spaCy
try:
    nlp = spacy.load("it_core_news_sm")
    print("Modello spaCy 'it_core_news_sm' caricato con successo.")
except OSError:
    print("Errore: Il modello 'it_core_news_sm' non è stato trovato.")
    nlp = None

if nlp:
    custom_stopwords = {
        "presidente", "legge", "onorevole","deputato", "articolo", "commissione",
        "perco", "disegno", "discussione", "proposta", "camera",
        "vicepresidente", "presidenza", "ordine", "pubblico", "sottoscritto",
        "votazione", "collega", "risposta", "relatore", "emendamento",
        "maggioranza", "senato", "italiano", "facoltare", "seduta", "comma", 
        "signore", "sottosegretario", "ministro", "parola", "provvedimento", 
        "problema", "candidato"
    }
    all_stopwords = set(STOP_WORDS).union(custom_stopwords)
    print(f"Set di stopwords creato. Dimensione totale: {len(all_stopwords)} parole.")

    def nlp_clean_docs(texts, batch_size=64):
        docs_out = []
        for doc in nlp.pipe(texts, batch_size=batch_size, disable=["parser","ner"]):
            toks = []
            for token in doc:
                if token.pos_ in {"NOUN", "ADJ"}:
                    lemma = token.lemma_.lower().strip()
                    if lemma and len(lemma) > 2 and lemma not in all_stopwords:
                        toks.append(lemma)
            docs_out.append(" ".join(toks))
        return docs_out

    nlp.max_length = 3_000_000 
    print(f"Limite di lunghezza di spaCy impostato a: {nlp.max_length} caratteri.")

    def quick_clean_text(s):
        s = str(s).lower()
        s = s.replace('\n', ' ').replace('\r', ' ')
        s = re.sub(r"[^a-zàèéìíîòóùúüäëïöÿßç\- ]+", " ", s)
        s = re.sub(r"\s+", " ", s).strip()
        return s

    def remove_speaker_name_completely(row):
        text = row['original_text']
        speaker = row['speaker']
        if pd.isna(text) or pd.isna(speaker) or not isinstance(speaker, str) or not speaker.strip():
            return text
        speaker_tokens = speaker.lower().split()
        for token in speaker_tokens:
            text = re.sub(r'\b' + re.escape(token) + r'\b', '', text, flags=re.IGNORECASE)
        text = text.replace(speaker.lower(), '')
        return re.sub(r'\s+', ' ', text).strip()

    print("\n--- Impostazione completata. Pronti per l'elaborazione. ---")
```

```python
# ===================================================================
# CELLA 2: APPLICAZIONE DELLA PIPELINE DI PULIZIA COMPLETA
# ===================================================================

if 'doc_df' in locals() and nlp:
    try:
        print("Inizio della pipeline di pulizia completa per la creazione di 'text_clean'...")

        print("Passo 1/3: Applicando pulizia di base (minuscolo, caratteri speciali)...")
        doc_df['temp_text'] = doc_df['text'].apply(quick_clean_text)

        print("Passo 2/3: Rimuovendo il nome del parlante dal testo pulito...")
        temp_df = doc_df[['speaker', 'temp_text']].rename(columns={'temp_text': 'original_text'}).copy()
        intermediate_texts = temp_df.apply(remove_speaker_name_completely, axis=1).tolist()
        
        print("Passo 3/3: Applicando lemmatizzazione e filtraggio con spaCy (potrebbe richiedere qualche minuto)...")
        doc_df['text_clean'] = nlp_clean_docs(intermediate_texts)

        doc_df.drop(columns=['temp_text'], inplace=True)

        print("\n--- Pipeline completata con successo! ---")
        print("La colonna 'text_clean' è stata creata e popolata per la rappresentazione dei topic.")

        print("\n--- Verifica Finale ---")
        try:
            example_doc_id_to_check = doc_id 
        except NameError:
            random_speech_for_check = doc_df.sample(n=1, random_state=101)
            example_doc_id_to_check = random_speech_for_check['doc_id'].iloc[0]

        example_row = doc_df[doc_df['doc_id'] == example_doc_id_to_check].iloc[0]
        print(f"Controllo per il Doc ID: {example_doc_id_to_check}")
        print(f"Parlante: {example_row['speaker']}")
        print("-" * 40)
        print("TESTO FINALE PULITO (colonna 'text_clean', pronto per CountVectorizer):")
        print(example_row['text_clean'])
        print("-" * 40)

        print("\nColonne attuali nel DataFrame:")
        print(doc_df.columns.tolist())

    except Exception as e:
        print(f"Si è verificato un errore durante l'elaborazione: {e}")

else:
    print("Errore: Il DataFrame 'doc_df' o l'oggetto 'nlp' non sono disponibili.")
    print("Per favore, esegui prima le celle di caricamento dati e di impostazione di spaCy.")
```

Calculate embeddings

Since most of the documents exceed the context window of the choosen embedding model (which has context window of 128 tokens), a chunking strategy is adopted. In this wasy, we avoid loosing information as it would happen adopting truncation.

```python

# Calculating embeddings

import pandas as pd
import numpy as np
from sentence_transformers import SentenceTransformer
from tqdm import tqdm
import re

# 2. CARICAMENTO  DATI
print("Fase 1/4: Caricamento del file CSV...")
try:
    doc_df = pd.read_csv(
        r"C:\Users\lisas\Desktop\ItaParl 1948-1972\new_id_texts.csv", 
        engine='python', 
        on_bad_lines='skip'
    )
    print("File caricato con successo.")
except Exception as e:
    print(f"ERRORE CRITICO: Impossibile leggere il file CSV. Errore: {e}")
    exit()

# check final quality
print("Fase 2/4: Rimozione righe con testo vuoto...")
initial_rows = len(doc_df)
doc_df.dropna(subset=['embedding_text'], inplace=True)
doc_df = doc_df[doc_df['embedding_text'].str.strip() != '']
doc_df.reset_index(drop=True, inplace=True)
final_rows = len(doc_df)
print(f"Rimosse {initial_rows - final_rows} righe vuote. Dimensioni finali: {doc_df.shape}")

# 4. embedding creation
print("Fase 3/4: Inizio creazione degli embedding (modalità RAM-friendly)...")

# Carica il modello
model = SentenceTransformer('paraphrase-multilingual-MiniLM-L12-v2')
print("Modello caricato.")

# chunking params
CHUNK_SIZE = 128
STRIDE = 64

# divide text in chunks
def split_into_chunks(text, chunk_size, stride):
    tokens = text.split()
    if len(tokens) <= chunk_size:
        return [text]
    chunks = []
    for i in range(0, len(tokens), stride):
        chunk = " ".join(tokens[i:i + chunk_size])
        chunks.append(chunk)
    return chunks

# list for collecting final embeddings for docs 
final_embeddings = []

# one doc at time
for text in tqdm(doc_df['embedding_text'], desc="Embedding dei documenti"):
    
    
    chunks = split_into_chunks(text, CHUNK_SIZE, STRIDE)
    
   
    if len(chunks) == 1:
        doc_embedding = model.encode(chunks[0], show_progress_bar=False)
        final_embeddings.append(doc_embedding)
        continue # Passa al documento successivo

    chunk_embeddings = model.encode(chunks, show_progress_bar=False)
    
   
    doc_embedding = np.mean(chunk_embeddings, axis=0)
    final_embeddings.append(doc_embedding)

# Add a new col in the df
doc_df['embeddings'] = final_embeddings

print("Embedding creati con successo!")

# Save
print("Fase 4/4: Salvataggio del DataFrame finale...")
output_path = r"C:\Users\lisas\Desktop\ItaParl 1948-1972\new_id_texts_with_embeddings.csv"
doc_df.to_csv(output_path, index=False, encoding='utf-8')

print(f"--- OPERAZIONE COMPLETATA! ---")
print(f"Il file finale è stato salvato in: {output_path}")
```

```python

# DEFINIZIONE DELLE FUNZIONI PER L'OTTIMIZZAZIONE


def create_bertopic_model(trial):
    """Crea un modello BERTopic con i parametri suggeriti da Optuna per un singolo trial."""
    # Parametri per UMAP
    umap_n_neighbors = trial.suggest_int('umap_n_neighbors', 10, 40)
    umap_n_components = trial.suggest_int('umap_n_components', 2, 8)
    umap_min_dist = trial.suggest_float('umap_min_dist', 0.0, 0.2)
    umap_metric = trial.suggest_categorical('umap_metric', ['euclidean', 'manhattan', 'cosine'])

    # Parametri per HDBSCAN
    hdbscan_min_cluster_size = trial.suggest_int('hdbscan_min_cluster_size', 10, 40)
    hdbscan_min_samples = trial.suggest_int('hdbscan_min_samples', 5, 20)
    hdbscan_metric = trial.suggest_categorical('hdbscan_metric', ['euclidean', 'manhattan'])
    hdbscan_cluster_selection_method = trial.suggest_categorical('hdbscan_cluster_selection_method', ['eom', 'leaf'])

    # Creazione dei componenti del modello
    
    embedding_model = SentenceTransformer('paraphrase-multilingual-MiniLM-L12-v2')
    
    umap_model = UMAP(
        n_neighbors=umap_n_neighbors,
        n_components=umap_n_components,
        min_dist=umap_min_dist,
        metric=umap_metric,
        random_state=42
    )

    hdbscan_model = HDBSCAN(
        min_cluster_size=hdbscan_min_cluster_size,
        min_samples=hdbscan_min_samples,
        metric=hdbscan_metric,
        cluster_selection_method=hdbscan_cluster_selection_method
    )

    # Creazione del modello BERTopic
    topic_model = BERTopic(
        embedding_model=embedding_model,
        umap_model=umap_model,
        hdbscan_model=hdbscan_model,
        verbose=False
    )
    return topic_model

def evaluate_model(topic_model, documents, embeddings):
    """Valuta il modello usando le metriche di Coerenza e Diversità."""
    try:
        # Addestra il modello
        topics, _ = topic_model.fit_transform(documents, embeddings)

        # Estrai le parole dei topic per il calcolo delle metriche
        topic_words_list = []
        for topic_id, words_scores in topic_model.get_topics().items():
            words = [word for word, score in words_scores]
            topic_words_list.append(words)

        output_for_octis = {'topics': topic_words_list}
        
        # Calcola la coerenza usando i documenti originali (già tokenizzati in text_clean)
        tokenized_docs = [doc.split() for doc in documents]
        coherence_metrics = Coherence(texts=tokenized_docs, measure='c_npmi')
        coherence = coherence_metrics.score(output_for_octis)

        # Calcola la diversità
        diversity_metric = TopicDiversity(topk=10)
        diversity = diversity_metric.score(output_for_octis)

        return coherence, diversity
    except Exception as e:
        print(f"Errore durante la valutazione: {e}")
        # Restituisce valori bassi in caso di errore per penalizzare il trial
        return -1.0, -1.0

def objective(trial):
    """Funzione obiettivo per l'ottimizzazione multi-obiettivo di Optuna."""
    # Crea il modello con i parametri del trial corrente
    topic_model = create_bertopic_model(trial)
    
    # Valuta il modello e restituisci le metriche
    coherence, diversity = evaluate_model(topic_model, documents, embeddings)
    
    return coherence, diversity
```

```python
def train_final_BERTopic_model(params, embeddings):
    """
    Addestra il modello BERTopic finale con i migliori parametri trovati.
    Valuta e restituisce il modello con le sue metriche finali.
    """

    # UMAP parameters
    umap_n_neighbors = params['umap_n_neighbors']
    umap_n_components = params['umap_n_components']
    umap_min_dist = params['umap_min_dist']
    umap_metric = params['umap_metric']

    # HDBSCAN parameters
    hdbscan_min_cluster_size = params['hdbscan_min_cluster_size']
    hdbscan_min_samples = params['hdbscan_min_samples']
    hdbscan_metric = params['hdbscan_metric']
    hdbscan_cluster_selection_method = params['hdbscan_cluster_selection_method']

    # Create components
    embedding_model = SentenceTransformer('paraphrase-multilingual-MiniLM-L12-v2') #prima era ('all-MiniLM-L6-v2')

    umap_model = UMAP(
        n_neighbors=umap_n_neighbors,
        n_components=umap_n_components,
        min_dist=umap_min_dist,
        metric=umap_metric,
        random_state=42
    )

    hdbscan_model = HDBSCAN(
        min_cluster_size=hdbscan_min_cluster_size,
        min_samples=hdbscan_min_samples,
        metric=hdbscan_metric,
        cluster_selection_method=hdbscan_cluster_selection_method,
        prediction_data=True
    )

    # Create BERTopic model
    # NOTA: Lasciamo che sia HDBSCAN a decidere il numero di topic per coerenza con l'ottimizzazione
    topic_model = BERTopic(
        embedding_model=embedding_model,
        umap_model=umap_model,
        hdbscan_model=hdbscan_model,
        verbose=False,
        calculate_probabilities=True,
        nr_topics = params['nr_topics']
    )

    # Fit the model on ALL your data
    topics, probs = topic_model.fit_transform(documents, embeddings)

    # --- VALUTAZIONE FINALE ---
    # Estrai le parole dei topic per calcolare le metriche
    topic_words_list = []
    for topic_id, words_scores in topic_model.get_topics().items():
        words = [word for word, score in words_scores]
        topic_words_list.append(words)

    output_for_octis = {'topics': topic_words_list}

    # CORREZIONE CRITICA: Usa 'documents' per calcolare la coerenza
    tokenized_docs = [doc.split() for doc in documents]

    coherence_metrics = Coherence(texts=tokenized_docs, measure='c_npmi')
    coherence = coherence_metrics.score(output_for_octis)

    diverisity_metric = TopicDiversity(topk=10)
    diversity = diverisity_metric.score(output_for_octis)

    return topic_model, coherence, diversity, topics, probs
```

```python
# Extracts 5 random documents for each topic from those with the highest assignment probability.
# This creates a sample file for qualitative topic analysis.

import pandas as pd
import numpy as np

# --- 1. Load the source Excel file ---
input_filename = r"C:\Users\lisas\Desktop\ItaParl 1948-1972\files finali\docs_with_topics_probs.xlsx"
df_wide = pd.read_excel(input_filename)

print("Excel file loaded. Starting restructuring...")

# --- 2. Transform data into a long format ---
data_long = []
for index, row in df_wide.iterrows():
    custom_id = row['custom_id']
    for i in range(1, 4):  # topic_1, topic_2, topic_3
        topic_col = f'topic_{i}'
        prob_col = f'prob_{i}'
        
        if pd.notna(row[topic_col]):
            topic_id_str = str(row[topic_col])
            topic_id_num = int(topic_id_str.split('_')[0])
            
            data_long.append({
                'custom_id': custom_id,
                'topic_id': topic_id_num,
                'probability': row[prob_col]
            })

df_long = pd.DataFrame(data_long)
print("Data restructured into long format.")

# --- 3. Group by topic and select top 5 documents ---
def create_topic_row(group):
    # Sort by probability in descending order
    group_sorted = group.sort_values('probability', ascending=False)
    
    top_docs = []
    
    # Select documents with the maximum probability
    max_prob = group_sorted['probability'].max()
    max_docs = group_sorted[group_sorted['probability'] == max_prob]
    
    if len(max_docs) > 5:
        # Randomly sample from the top candidates, reproducible with random_state
        top_docs = max_docs.sample(5, random_state=42)
    else:
        top_docs = max_docs
        # If fewer than 5, add other documents with lower probability
        if len(top_docs) < 5:
            rest = group_sorted[group_sorted['probability'] < max_prob]
            top_docs = pd.concat([top_docs, rest.head(5 - len(top_docs))])
    
    new_row_data = {'ID_TOPIC': group.name}
    
    # Populate docX / probX columns
    for i, (_, row) in enumerate(top_docs.iterrows()):
        new_row_data[f'doc{i+1}'] = row['custom_id']
        new_row_data[f'prob{i+1}'] = row['probability']
        
    return new_row_data

final_df = df_long.groupby('topic_id').apply(create_topic_row).reset_index(drop=True)

print("Final DataFrame created with the desired structure.")

# --- 4. Save the result to a new Excel file ---
output_filename = r"C:\Users\lisas\Desktop\ItaParl 1948-1972\files finali\top5_docs_per_topic_random.xlsx"
final_df.to_excel(output_filename, index=False, engine='openpyxl')

print(f"\n✅ Excel file '{output_filename}' created successfully!")
print("Structure: ROWS = Topic, COLUMNS = doc1, prob1, ..., doc5, prob5")

# Preview
print("\nPreview of the created DataFrame:")
print(final_df.head())
```
